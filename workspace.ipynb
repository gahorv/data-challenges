```python
def odds(x):
    
    try:
        x = float(x)
    except: 
        x = x.replace("-","1")
        x = int(x)
        
    return x
```

```python
m_o_df["odds"] = m_o_df["odds"].apply(odds)
```

```python
#largest-odds-overcome-in-game
m_o_df["score_home"], m_o_df["score_away"] = m_o_df["score"].str.split('-', 1).str
m_o_df["score_home"] = m_o_df["score_home"].apply(int)
m_o_df["score_away"] = m_o_df["score_away"].apply(int)


conditions = [
    (m_o_df['score_home'] > m_o_df['score_away']),
    (m_o_df['score_home'] < m_o_df['score_away']),
    (m_o_df['score_home'] == m_o_df['score_away'])]
choices = [1, 2, 0]
m_o_df['eredmeny'] = np.select(conditions, choices)
m_o_df

m_o_df.loc[(m_o_df['eredmeny'] != 0)]['odds'].max()
```
